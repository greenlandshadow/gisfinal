```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import geopandas as gpd
import contextily as ctx

# 假设gdf是您的GeoDataFrame，其中包含年份和位置数据
# ...

# 确保数据使用的是Web Mercator投影
gdf = gdf.to_crs(epsg=3857)

# 获取年份范围
years = range(2010, 2023)

# 创建绘图
fig, ax = plt.subplots(figsize=(10, 10))

# 设定固定的边界范围
xmin, ymin, xmax, ymax = gdf.total_bounds
dx, dy = xmax - xmin, ymax - ymin
# 增加边界范围的一点余量
padding_factor = 0.1
xlim = (xmin - padding_factor * dx, xmax + padding_factor * dx)
ylim = (ymin - padding_factor * dy, ymax + padding_factor * dy)

def animate(year):
    ax.clear()  # 清除之前的帧
    gdf_year = gdf[gdf['year'] == year]
    sample_count = len(gdf_year)  # 获取样本数
    gdf_year.plot(ax=ax, color='blue', markersize=7)
    
    # 设置标题，包括年份和样本数
    ax.set_title(f"YEAR: {year} - SAMPLE_COUNT: {sample_count}", fontsize=15)

    # 使用固定的边界范围
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)

    # 添加底图，并设置固定的缩放级别
    ctx.add_basemap(ax, source=ctx.providers.OpenStreetMap.Mapnik, zoom=12)

# 创建动画
ani = animation.FuncAnimation(fig, animate, frames=years, interval=2000)  # 每帧显示2秒

# 保存动画
ani.save('BEIJING LAND.gif', writer='imagemagick', fps=1)  # 设置每秒0.5帧，即每帧2秒

plt.close()
```
